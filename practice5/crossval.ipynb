```python

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from sklearn.linear_model import LogisticRegression
#from sklearn.

data=pd.read_csv('C:/Users/Virag/Desktop/Machine Learning/practice5/data/heart_disease.csv')
labels=data.values[:,-1]
labels[labels>1]=1
labels=labels.astype(int)

data=data.values[:,:-1]

lr=LogisticRegression(max_iter=10000)
```

```python
lr.fit(data,labels)
lr.predict(data)
lr.score(data,labels)
```

```python
# Cross-validation
from sklearn.model_selection import cross_validate
k_fold = [3,5,7,10,20,50]

training_mean = []
testing_mean = []
training_dev = []
testing_dev = []
for k in k_fold:
    cross = cross_validate(lr, data, labels, cv=k, return_train_score=True)
    train_mean = np.mean(cross['train_score'])
    train_dev = np.std(cross['train_score'])
    test_mean = np.mean(cross['test_score'])
    test_dev = np.std(cross['test_score'])
    training_mean.append(train_mean)
    training_dev.append(train_dev)
    testing_mean.append(test_mean)
    testing_dev.append(test_dev)
```

```python
import matplotlib.pyplot as plt
plt.plot(training_mean,label='Train mean values')
plt.plot(testing_mean,label='Test mean values')
plt.legend(loc='upper right')
```

```python
plt.plot(training_dev,label='Train deviation values')
plt.plot(testing_dev,label='Test deviation values')
plt.legend(loc='upper left')
```
