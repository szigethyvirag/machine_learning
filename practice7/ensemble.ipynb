```python

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from sklearn.linear_model import LogisticRegression
#from sklearn.

data=pd.read_csv('../data/heart_disease.csv')
labels=data.values[:,-1]
labels[labels>1]=1
labels=labels.astype(int)

data=data.values[:,:-1]
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier(max_depth=None, min_samples_split=2, random_state=0)
score = cross_val_score(clf, data, labels, cv=5)
score.mean()
```

Blending:

```python
from sklearn.ensemble import StackingClassifier
import time

trees = []
for i in range(50):
    tmp = DecisionTreeClassifier(max_depth=None, min_samples_split=2,random_state=0)
    trees.append((f"{i} :(",tmp))

clf = StackingClassifier(estimators=trees, final_estimator=LogisticRegression())
start = time.time()
score = cross_val_score(clf, data, labels, cv=5)
print(f"Time during the train and test: {time.time()-start}")
print(f"Mean: {score.mean()}, Standard deviation: {score.std()}")

```

Bagging:

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_estimators = 50, max_depth=2, random_state=0)
start = time.time()
score = cross_val_score(clf, data, labels, cv=5)
print(f"Time during the train and test: {time.time()-start}")
print(f"Mean: {score.mean()}, Standard deviation: {score.std()}")
```

Boosting:

```python
from sklearn.ensemble import AdaBoostClassifier

clf = AdaBoostClassifier(n_estimators = 50, random_state=0)
start = time.time()
score = cross_val_score(clf, data, labels, cv=5)
print(f"Time during the train and test: {time.time()-start}")
print(f"Mean: {score.mean()}, Standard deviation: {score.std()}")
```

I would use different ensemble methods if my goal would be to achieve better performances in: 
Memory: Boost (because of theoretical assumptions -> it does not need every tree at the same time)
Speed: Blend
Performance: Bag
